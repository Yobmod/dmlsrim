```python
"""create fig with 1 ax
add axis titles

create ax
create legend

add ax to fig
update legend"""


import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
```

```python
# create data

x = [1,2,3,4,5,6,7,8,9]
y0 = [2,3,5,6,8,9,6,35,36]
y1 = [0,4,8,9,16,20, 18, 13, 4]
assert len(x) == len(y0) == len(y1)
```

```python

def create_fig():
    fig, ax = plt.subplots()
    ax.set_xlabel(f'Depth []')
    ax.set_ylabel(f'Collision damage ')
    fig.suptitle('Damage Energy vs. Depth', fontsize=15)
    fig.set_size_inches((10, 6))
    return fig, ax

def plot_axes(data_x, data_y, label, ax=None):
    if ax is None:
        ax = plt.gca()
    ax.plot(data_x, data_y, label=label)
    ax.legend()
    return ax


f, ax = create_fig()
plot_axes(x, y0, "y0", ax)
plot_axes(x, y1, "y1", ax)
```
